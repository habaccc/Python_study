# 흡연 여부에 따른 tip의 차이?

*  흡연 여부(Yes/No) 별 tip의 평균

```python
tip_mean_by_smoker = tips.groupby('smoker')['tip'].mean()
print(f'흡연 tip 평균: {tip_mean_by_smoker["Yes"]}')
print(f'비흡연 tip 평균: {tip_mean_by_smoker["No"]}')
```

# 요일에 따른 tip의 차이?

*  요일 별 tip의 평균


```python
tip_mean_by_day = tips.groupby('day')['tip'].mean()
print(f'목요일 tip 평균: {tip_mean_by_day["Thur"]}')
print(f'금요일 tip 평균: {tip_mean_by_day["Fri"]}')
print(f'토요일 tip 평균: {tip_mean_by_day["Sat"]}')
print(f'일요일 tip 평균: {tip_mean_by_day["Sun"]}')
```

# 시간에 따른 tip의 차이?

*  시간 별 tip의 평균


```python
tip_mean_by_time = tips.groupby('time')['tip'].mean()
print(f'점심 tip 평균: {tip_mean_by_time["Lunch"]}')
print(f'저녁 tip 평균: {tip_mean_by_time["Dinner"]}')
```
