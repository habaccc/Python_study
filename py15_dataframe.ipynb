# 성별에 따른 tip 차이?

*  남성 손님들의 tip의 평균, 여성 손님들의 tip의 평균
*  성별 tip의 최댓값
*  성별 tip의 최솟값

```python
genders = tips['sex'].unique() # distinct와 비슷
```

```python
for x in tips.sex.unique():
  # print(x)
  tip_mean = tips[tips.sex == x].tip.mean()
  print(f'{x}: {tip_mean}')
```

```python
tips[tips.sex == 'Female'].tip.mean()
print(f'{x}: {tip_mean}')
```

```python
for x in genders:
  tip_max = tips[tips.sex == x].tip.max()
  print(f'{x}: {tip_max}')
```

```python
tip_mean = tips.groupby('sex')['tip'].mean()
print(f'남성 손님들의 tip 평균: {tip_mean["Male"]}')
print(f'여성 손님들의 tip 평균: {tip_mean["Female"]}')
print('----------------')

tip_max = tips.groupby('sex')['tip'].max()
print(f'남성 tip의 최댓값: {tip_max["Male"]}')
print(f'여성 tip의 최댓값: {tip_max["Female"]}')
print('----------------')

tip_min = tips.groupby('sex')['tip'].min()
print(f'남성 tip의 최솟값: {tip_min["Male"]}')
print(f'여성 tip의 최솟값: {tip_min["Female"]}')
```

# 흡연 여부에 따른 tip의 차이?

*  흡연 여부(Yes/No) 별 tip의 평균

```python
for x in tips.smoker.unique():
  tip_mean = tips[tips.smoker == x]
  print(f'smoker {x}: {tip_mean}')
```

```python
tip_mean_by_smoker = tips.groupby('smoker').tip.mean()
print(f'흡연 tip 평균: {tip_mean_by_smoker["Yes"]}')
print(f'비흡연 tip 평균: {tip_mean_by_smoker["No"]}')
```

# 요일에 따른 tip의 차이?

*  요일 별 tip의 평균


```python
for x in tips.day.unique():
  tip_mean = tips[tips.day == x].tip.mean()
  print(f'{x}: {tip_mean}')
```

```python
tip_mean_by_day = tips.groupby('day')['tip'].mean()
print(f'목요일 tip 평균: {tip_mean_by_day["Thur"]}')
print(f'금요일 tip 평균: {tip_mean_by_day["Fri"]}')
print(f'토요일 tip 평균: {tip_mean_by_day["Sat"]}')
print(f'일요일 tip 평균: {tip_mean_by_day["Sun"]}')
```

카테고리 별 2개 이상 변수(컬럼)의 통계값 찾기

```python
# 성별 영수증, 팁의 평균
tips.groupby('sex')[['total_bill', 'tip']].mean()
```

```python
# 성별, 흡연여부별 영수증과 팁의 평균
tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].mean()
```

## 그룹별 2개 이상의 통계 함수 적용

```python
# 성별 팁의 평균, 최솟값, 중앙값, 최댓값
tips.groupby('sex').tip.agg([np.mean, np.min, np.median, np.max]) # aggregate
```

```python
# 성별, 흡연여부별 영수증과 팁의 평균과 표준편차
tips.groupby(['sex','smoker'])[['total_bill', 'tip']].agg([np.mean, np.std])
```

## pivot_table함수(메서드)

* Parameters:
  * values: 통계 함수를 적용할 변수(컬럼)
  * index: 결과 데이터프레임에서 행 인덱스로 사용할 변수(컬럼)
  * columns: 결과 데이터프레임에서 열 인덱스로 사용할 변수(컬럼)
  * aggfunc: aggregation function. 기본값은 평균(mean).

```python
# 성별 팁의 평균
tips.pivot_table(values='tip', index='sex') # aggfunc='mean' (기본값)
```

```python
# 성별 흡연여부별 팁의 평균
tips.pivot_table(values='tip', index=['sex', 'smoker'])
```

```python
tips.pivot_table(values='tip', index='sex', columns='smoker')
```

```python
# 성별, 흡연여부별, 요일별, 시간별 팁의 평균
tips.pivot_table(values='tip', index=['sex', 'smoker'], columns=['day', 'time'])
```

```python
# 요일별, 시간별 영수증과 팁의 평균, 표준편차
tips.pivot_table(values=['total_bill','tip'], index=['day', 'time'], aggfunc=[np.mean, np.std])
```

```python
tips.pivot_table(values=['total_bill', 'tip'], index='day', columns='time', aggfunc=[np.mean, np.std])
```
